```python
x=Bank.drop('Personal Loan',axis=1)
y=Bank['Personal Loan']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3,random_state=100,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
sc.fit(xtrain) # found the metrics mean and sd
```

```python
xtrain=pd.DataFrame(sc.fit_transform(xtrain),columns=xtrain.columns)
```

```python
xtest.head()
```

```python
xtest=pd.DataFrame(sc.transform(xtest),columns=xtest.columns)
```

```python
xtest.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression(random_state=100)
```

```python
model_LR=LR.fit(xtrain,ytrain).predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report,f1_score,cohen_kappa_score,mean_squared_error
```

```python
accuracy_score(ytest,model_LR)
```

```python
from sklearn import metrics
f1_logReg=metrics.f1_score(ytest,model_LR,average='weighted')
f1_logReg
```

```python
print(classification_report(ytest,model_LR))
```

```python
print('kappa_score:',metrics.cohen_kappa_score(ytest,model_LR))
kappa_logReg=metrics.cohen_kappa_score(ytest,model_LR)
```

```python
print('Accuracy:',metrics.accuracy_score(ytest,model_LR))
acc_logReg=metrics.accuracy_score(ytest,model_LR)
fpr,tpr,_=metrics.roc_curve(ytest,model_LR)
auc_logReg=metrics.auc(fpr,tpr)
auc_logReg
```

```python
from sklearn import metrics
print('F1 score:',metrics.f1_score(ytest,model_LR,average='weighted'))
```

```python
from sklearn.model_selection import KFold
from sklearn import metrics
kf=KFold(n_splits=5,shuffle=True,random_state=100)
root=[]
lst=[]
for train,test in kf.split(x,y):
    logReg=LogisticRegression()
    xtrain,xtest=x.iloc[train,:],x.iloc[test,:]
    ytrain,ytest=y.iloc[train],y.iloc[test]
    ypred_logReg=logReg.fit(xtrain,ytrain).predict(xtest)
    root.append(metrics.accuracy_score(ytest,ypred_logReg))
    fpr,tpr,_=metrics.roc_curve(ytest,ypred_logReg)
    lst.append(metrics.auc(fpr,tpr))
```

```python
np.mean(root) # mean accuracy in kfold
```

```python
np.mean(lst) # mean auc in kfold
```

```python
z=pd.DataFrame({'Model':['Logistic'],'F1_score':f1_logReg,'kappa_score':kappa_logReg,'Accuracy':acc_logReg,'Auc':auc_logReg})
z = z[['Model', 'F1_score', 'kappa_score','Accuracy','Auc']]
z
```

```python
# kmeans clustering
from sklearn.cluster import KMeans
cluster_range = range( 1,26) 
cluster_errors = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 5)
  clusters.fit(Bank)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
clusters_df[0:20]
```

```python
# elbow plot
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```
